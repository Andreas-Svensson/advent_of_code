using python, read a file called "assets/1.txt" and split its contents on newline, then store to a variable called "calories"

```python
# Open the file in read mode
with open("assets/1.txt", "r") as file:
  # Read the contents of the file
  contents = file.read()

# Split the contents on newline
calories = contents.split("\n")
```

using python, write a script that takes a list of strings called "calories" and converts them to integers where possible, keeping empty strings intact

```python
# Define a function that takes a list of strings
def convert_to_ints(calories):
  # Create a new empty list
  converted_calories = []

  # Iterate over the list of strings
  for cal in calories:
    # If the string is empty, append it to the new list as-is
    if cal == "":
      converted_calories.append(cal)
    # Otherwise, try to convert the string to an integer and append it to the new list
    else:
      try:
        converted_calories.append(int(cal))
      except ValueError:
        # If the conversion fails, print an error message and append the original string to the new list
        print(f"Could not convert {cal} to an integer.")
        converted_calories.append(cal)

  # Return the new list
  return converted_calories

# Call the function with the list of strings
converted_calories = convert_to_ints(calories)
```

where the input is an already existing variable named "converted_calories" which is a list containing integers and empty strings denoting a new elf

```python
# initialize variables to keep track of the current and overall max calories
current_max = 0
overall_max = 0

# iterate over the list of converted_calories
for calorie in converted_calories:
  # if the calorie is an empty string, reset the current max
  if calorie == "":
    current_max = 0
  else:
    # add the current calorie to the current max
    current_max += calorie
    # if the current max is greater than the overall max, update the overall max
    if current_max > overall_max:
      overall_max = current_max

# print the overall max
print(overall_max)
```
